# Job Reassigment Problem en QAOA: Introduccion

Aca se definen las variables de la instancia que se usaran para el QUBO.Se definiran como parametros a traves de estructuras de datos de Python. Los parametros son:

* priorities: las prioridades que puede tener cada trabajo. A menor el valor, mayor prioridad tiene el trabajo.
* affinities: las afinidades que puede tener cada agente con un trabajo en particular. A menor el valor, mayor la afinidad del agente con un trabajo.
  
* agents: lista de los agentes
* agents_assgJobs: relacion 1a1. El trabajo asignado al agente $\textit{i}$ se encuentra en indice $\textit{i}$ de la lista.
* agents_assgJobsAfinnity: relacion 1aM. En el indice $\textit{i}$ se encuentra la lista de afinidades para cada uno de los trabajos ya asignados, donde en el indice $\textit{j}$ de esta sublista se muestra la afinidad particular al trabajo asignado $\textit{j}$.
* agents_vacnJobsAfinnity: relacion 1aM. En el indice $\textit{i}$ se encuentra la lista de afinidades para cada uno de los trabajos vacantes, donde en el indice $\textit{j}$ de esta sublista se muestra la afinidad particular al trabajo vacante $\textit{j}$.

* assgJobs: lista de trabajos asignados
* assgJobs_agents: relacion 1a1. El agente que esta en el trabajo asignado $\textit{i}$ se encuentra en indice $\textit{i}$ de la lista. Lista inversa a la agents_assgJobs.
* assgJobs_priority: relacion 1a1. La priorioridad del trabajo asignado $\textit{i}$ se encuentra en indice $\textit{i}$ de la lista.

* vacnJobs: lista de trabajos vacantes
* vacnJobs_priority: relacion 1a1. La priorioridad del trabajo vacante $\textit{i}$ se encuentra en indice $\textit{i}$ de la lista.

* priorityWeightCoeff: coeficiente que marca que tanto efecto tendra en el costo final el calculo de la prioridad.
* affinityWeightCoeff:coeficiente que marca que tanto efecto tendra en el costo final el calculo de la afinidad.

* penalty1 = penalty asociado al 1er constraint (ver paper)
* penalty2 =  penalty asociado al 2do constraint (ver paper)

* allBinaryVariable: lista de todas las variables booleana que se usaran. Se define como el producto entre la cantidad de agentes y de trabajos vacantes.

De las siguientes celdas de codigo de esta seccion, debe ejecutarse unicamente aquella que se quiera trabajar a lo largo del notebook. La celda ejecuta representa la instancia de problema que se convertira a QUBO y que sera usada en el algoritmo QAOA.

### 1.2 Algoritmo de formulacion QUBO

Este algoritmo toma las variable de la celda de instancia anteriormente ejecutada y crea la formulación QUBO correspondiente. En terminos de salida de codigo, se crea un diccionario $\texttt{terms\_weights}$. Cada elemento del diccionario puede interpretarse como un elemento $w* x_i * x_j$  tal que $x_i,x_j\in[0,1],w \in R$. La clave de cada elemento del diccionario sera un entero(o tupla de enteros) representando su variable(s) $x_i$(o $x_i *x_j$). El valor de cada elemento del diccionario sera un real representando al respectivo $w$.

```python
terms_weights = {}
terms = []
weights = []

for i in range(len(vacnJobs)):
    for j in range(len(agents)):        
        finalPriority = priorityWeightCoeff * (vacnJobs_priority[i] - assgJobs_priority[agents_assgJobs[j]])
        finalAffinity = affinityWeightCoeff * (agents_vacnJobsAfinnity[j][i] - agents_assgJobsAfinnity[j][agents_assgJobs[j]])
        finalWeight = -1*(finalPriority + finalAffinity)

        terms_weights[(allBinaryVariables[len(agents)*i+j])] = finalWeight

print('----------------------------------------')
print('After 1 variable terms')
print('terms:',list(terms_weights.keys()))
print('weights: ',list(terms_weights.values()))
print('terms_weights: ',terms_weights)
print()


#this aux_weights will have the terms with two variables. Finally, its elements will be appended to the original terms list.
auxWeights = []
for i in range(len(vacnJobs)):
    for j in range(len(agents)):
        terms_weights[len(agents)*i+j] = terms_weights[len(agents)*i+j] + penalty1*(1 - 2 * 0.5)
        for r in range(j+1,len(agents)):
            terms_weights[(allBinaryVariables[len(agents)*i+j],allBinaryVariables[len(agents)*i+r])] = 2*penalty1

print('----------------------------------------')
print('After first penalty')
print('terms:',list(terms_weights.keys()))
print('weights: ',list(terms_weights.values()))
print('terms_weights: ',terms_weights)
print()


auxWeights = []
for j in range(len(agents)):
    for i in range(len(vacnJobs)):
        terms_weights[len(vacnJobs)*j+i] = terms_weights[len(vacnJobs)*j+i] + penalty2*(1 - 2 * 0.5)
        for r in range(i+1,len(vacnJobs)):
            key = (allBinaryVariables[len(agents)*i+j],allBinaryVariables[len(agents)*r+j])
            if key in terms_weights:
                terms_weights[(allBinaryVariables[len(agents)*i+j],allBinaryVariables[len(agents)*r+j])] += 2*penalty2
            else:
                terms_weights[(allBinaryVariables[len(agents)*i+j],allBinaryVariables[len(agents)*r+j])] = 2*penalty2

print('----------------------------------------')
print('After second penalty')
print('terms:',list(terms_weights.keys()))
print('weights: ',list(terms_weights.values()))
print('terms_weights: ',terms_weights)
print()
```

Adaptamos el formato de los terms y weights calculados en el algoritmo anterior para que sea compatible con las librerias $\textit{qiskit optimization}$ y $\textit{OpenQAOA}$.

```python
terms = list(terms_weights.keys())
weights = list(terms_weights.values())

terms_aux = []
for term in terms:
    if type(term) == tuple:
        terms_aux.append([term[0],term[1]])
    else:
        terms_aux.append([term])
terms = terms_aux
print('final terms: ',terms)
print('final weights: ',weights)
        
```

### 1.3 Uso del QUBO creado en Qiskit Optimization

```python
linear = {clave: terms_weights[clave] for clave in list(terms_weights.keys())[:12]}
quad = {clave: terms_weights[clave] for clave in list(terms_weights.keys())[12:]}
print(linear)
print(quad)

mod = QuadraticProgram("JRP")
for i in allBinaryVariables:
    mod.binary_var(name="x"+str(i))

mod.minimize(constant=3, linear=linear, quadratic=quad)
print(mod.prettyprint())
result = CplexOptimizer().solve(mod)
print(result.prettyprint())

resultstring =""
for val in result.variables_dict.values():
    resultstring += str(int(val))
print('solucion: ',resultstring)
```

### 1.3 Uso de QAOA con el QUBO creado

```python
# se convierte el QUBO su equivalente de formulacion Ising
ising_terms,ising_weights = QUBO.convert_qubo_to_ising(len(allBinaryVariables), terms, weights)

# se crea el qubo Ising
qubo = QUBO(n = len(allBinaryVariables), terms=ising_terms, weights=ising_weights)
display(qubo.hamiltonian.expression)

# se resuelve usando algoritmo clasico de fuerza bruta
a = QAOA()
a.compile(qubo)
a.solve_brute_force(verbose=True)
from openqaoa.utilities import ground_state_hamiltonian
#energy, configuration = ground_state_hamiltonian(qubo.hamiltonian)
#print(f"Ground State energy: {energy}, Solutions: {configuration}")
#display(qubo.asdict())
```

### 1.5 Discuciones

En cuanto a las instancias 1,2 y 4, se obtienen muy bueno resultados, sobretodo teniendo en cuenta que no hay una gran ingenieria detras de los parametros de QAOA configurados. Los parametros que mas cambian la probabilidad de exito del algoritmo, no son del algoritmo perse, sino de la funcion QUBO. El valor de los penalties y de los pesos afinidad y prioridad causan que un algoritmo pase de un rendimiento mediocre a uno muy bueno y viceversa. Se invita a probar experimentar con estos parametros en las diferentes instancias del problema.

Como futuras pruebas, seria interesante detectar patrones de comportamiento en lo que respecta a los penalties y los pesos de afinidad y prioridad. Si se encuentra como automatizar la configuracion de estos valores, podria lograrse que el algoritmo tenga un buen desempeño indiferentemente de la instancia del problema.

Ya teniendo una buena configuracion de penalties y pesos afinidad/prioridad, el otro parametro que genera un cambio es la cantidad de capas del circuito QAOA $p$. Utilizando la configuracion del QUBO de cada instancia que vienen por defecto en el notebook (son las mejores que se encontro), experimentar variar el valor de $p$ entre 5 a 15. Se vera un incremento de probabilidad de exito al ir aumentando a $p$.
